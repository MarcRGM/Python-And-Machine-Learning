# NumPy Vector Explorer

This notebook introduces **NumPy**, the foundational library for numerical computing in Python.

## Environment Setup

NumPy is imported under the standard alias `np`. This alias is used universally in the Python data ecosystem.

```python
import numpy as np
```

## Creating Arrays from Data
The `dtype=float` argument is important: it forces the array to a floating‑point type so that `None` values become `np.nan` (Not a Number). NumPy's statistical functions understand `np.nan` and can skip it.

```python
data = {
    "age": [25, 30, 35, None, 40, 120, 29, 31],
    "income": [50000, 60000, 75000, 80000, None, 95000, 62000, 58000],
    "score": [85, 90, 78, 92, 88, None, 76, 95]
}

np_data = {}
for col in data:
    np_data[col] = np.array(data[col], dtype=float)
```

---
## 1. Array Summary Function

`array_summary(arr)` takes a one‑dimensional NumPy array and returns a dictionary of descriptive statistics. It uses NumPy's `nan*` functions, which compute each statistic while ignoring `NaN` values.

### Functions used
- `np.nanmean()` – mean (average), ignoring NaN.
- `np.nanmedian()` – median (middle value when sorted), ignoring NaN.
- `np.nanstd()` – standard deviation (spread), ignoring NaN.
- `np.nanmin()` and `np.nanmax()` – smallest and largest values, ignoring NaN.
- `np.isnan().sum()` – counts how many NaN values exist.

```python
def array_summary(arr):
    """Return a dictionary of summary statistics for a 1D NumPy array."""
    summary = {}
    summary["mean"] = np.nanmean(arr)
    summary["median"] = np.nanmedian(arr)
    summary["std"] = np.nanstd(arr)
    summary["min"] = np.nanmin(arr)
    summary["max"] = np.nanmax(arr)
    summary["missing_count"] = np.isnan(arr).sum()
    return summary
```

The function is called on each array.

```python
for col in np_data:
    print(array_summary(np_data[col]))
```

## 2. Vectorized Age Comparison

`compare_ages(age)` demonstrates **vectorized arithmetic**: the expression `age + 10` adds 10 to every element of the array in one step. No loop is needed.

This is the key advantage NumPy provides over plain Python lists. The operation runs in compiled C code, making it much faster for large datasets.

```python
def compare_ages(age):
    """Return the mean of current ages and ages shifted by 10 years."""
    current_age_mean = np.nanmean(age)
    future_age_mean = np.nanmean(age + 10)
    return current_age_mean, future_age_mean
```

```python
current, future = compare_ages(age)
print(f"Current mean age: {current:.2f}")
print(f"Future mean age: {future:.2f}")
```
