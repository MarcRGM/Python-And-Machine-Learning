### Data Summary Tool
#### This notebook demonstrates basic **Exploratory Data Analysis (EDA)** using Pandas and Matplotlib. It computes descriptive statistics and plots histograms for each numeric column.


### **Import libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### **Sample Data**
#### Creating a mock dataset with numeric columns and some missing values.

```python
data = {
    "age": [25, 30, 35, None, 40, 120, 29, 31],
    "income": [50000, 60000, 75000, 80000, None, 95000, 62000, 58000],
    "score": [85, 90, 78, 92, 88, None, 76, 95]
}
df = pd.DataFrame(data)
```

### **1. Summary Statistics Function**
#### This function returns a dictionary of statistics **without** plotting.

```python
def summarize_data(df):
    """Return a dictionary with mean, median, std, min, max, missing_values for each numeric column."""
    col_stats = {}
    for col in df.columns:
        stats = {}
        if pd.api.types.is_numeric_dtype(df[col]):
            stats['mean'] = df[col].mean()
            stats['median'] = df[col].median()
            stats['std'] = df[col].std()
            stats['min'] = df[col].min()
            stats['max'] = df[col].max()
            stats['missing_values'] = df[col].isnull().sum()
            col_stats[col] = stats
    return col_stats

print(summarize_data(df))
```

### **2. Histogram Plotting Function**
#### A separate function that only handles visualization.

```python
def plot_histogram(df, bins=4):
    """Plot a histogram for each numeric column after dropping missing values."""
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            plt.hist(df[col].dropna(), bins=4)
            plt.title(f"{col} distribution")
            plt.xlabel(f"{col}")
            plt.ylabel("frequency")
            plt.show()
    return None

plot_histogram(df)
```
