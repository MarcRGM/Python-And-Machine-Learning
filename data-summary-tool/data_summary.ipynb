# Data Summary Tool

This notebook demonstrates basic **Exploratory Data Analysis (EDA)** using **Pandas** and **Matplotlib**. It computes descriptive statistics and plots histograms for each numeric column.

Two core libraries are used:
- **pandas** – provides DataFrames and functions to load, clean, and summarise tabular data.
- **matplotlib.pyplot** – a plotting library for creating charts and histograms.

## Environment Setup

The libraries are imported first. The aliases `pd` and `plt` are community standards that make code shorter and more recognisable.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Creating a Sample Dataset

A small dictionary is created. Pandas converts it into a **DataFrame** - a two‑dimensional, tabular data structure similar to a spreadsheet or SQL table. Each key becomes a column name, and the list values become the column entries.

The dataset has three numeric columns (`age`, `income`, `score`) and intentionally contains some `None` values to simulate real‑world missing data.

```python
data = {
    "age": [25, 30, 35, None, 40, 120, 29, 31],
    "income": [50000, 60000, 75000, 80000, None, 95000, 62000, 58000],
    "score": [85, 90, 78, 92, 88, None, 76, 95]
}
df = pd.DataFrame(data)
```

The `None` entries are displayed as `NaN` ("Not a Number"). In pandas, this is a standard marker for missing data. `NaN` are ignored by default in mathematical functions, but visualisations often break.

---
## 1. Summary Statistics Function

`summarize_data(df)` extracts the most important descriptive statistics for every numeric column and returns them as a dictionary.

### How the function works

1. **Loop through columns**  
   `for col in df.columns:` iterates over every column name in the DataFrame.

2. **Check if numeric**  
   `pd.api.types.is_numeric_dtype(df[col])` returns `True` if the column contains numbers (integers, floats). Non‑numeric columns like text or dates are skipped to avoid calculation errors.

3. **Compute statistics** for each numeric column:
   - `mean()` – the average value (sum divided by count). Sensitive to outliers.
   - `median()` – the middle value when sorted. Resilient against outliers.
   - `std()` – **standard deviation**, a measure of spread. A low std means values are clustered near the mean; a high std means they are widely scattered.
   - `min()` and `max()` – the smallest and largest values in the column.
   - `isnull().sum()` – `isnull()` returns a boolean filter (True where missing), then `sum()` counts the Trues, giving the total number of missing entries.

4. **Store in a nested dictionary**  
   Each column name becomes a key, mapping to a dictionary of its statistics.

```python
def summarize_data(df):
    """Return a dictionary with descriptive statistics for each numeric column."""
    col_stats = {}
    for col in df.columns:
        stats = {}
        if pd.api.types.is_numeric_dtype(df[col]):
            stats['mean'] = df[col].mean()
            stats['median'] = df[col].median()
            stats['std'] = df[col].std()
            stats['min'] = df[col].min()
            stats['max'] = df[col].max()
            stats['missing_values'] = df[col].isnull().sum()
            col_stats[col] = stats
    return col_stats
```

The function is run and the output shown.

```python
stats = summarize_data(df)
```

### Interpreting the `age` column

| Statistic | Value | What it tells us |
|-----------|-------|------------------|
| `mean` | ~44.3 | The average is pulled upward by the 120‑year‑old outlier |
| `median` | 31.0 | Half the ages are below 31 |
| `std` | ~33.2 | High spread, values vary widely from the mean |
| `min` / `max` | 25 / 120 | The range reveals the extreme outlier |
| `missing_values` | 1 | One row lacks an age entry |

When `mean` and `median` are far apart, the data is skewed or contains outliers. The `median` is often the safer summary.

## 2. Histogram Plotting Function

`plot_histograms(df, bins=4)` draws one histogram per numeric column.

### How the function works

1. **Loop through columns** and check for numeric type.
2. **Drop missing values** with `dropna()`. The `hist()` function cannot process `NaN`, it would crash. Dropping ensures only real numbers are plotted.
3. **Create a histogram** with `plt.hist()`. The `bins` parameter splits the data range into equal‑width intervals, counting how many values fall into each.
4. **Add labels** (`title`, `xlabel`, `ylabel`) so the chart is self‑explanatory.
5. **Display the plot** with `plt.show()`. This opens a window (or renders inline in a notebook).

```python
def plot_histograms(df, bins=4):
    """Plot one histogram per numeric column after dropping missing values."""
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            plt.hist(df[col].dropna(), bins=bins)
            plt.title(f"{col} distribution")
            plt.xlabel(f"{col}")
            plt.ylabel("frequency")
            plt.show()
    return None
```

The plotting function is called to see the distributions.

```python
plot_histograms(df, bins=4)
```

### Reading the `age` histogram

Most bars cluster between 20 and 40. A lone bar sits far to the right near 120. This is a classic **right‑skewed distribution** - the long tail pulls the mean upward while the median stays in the main cluster.
